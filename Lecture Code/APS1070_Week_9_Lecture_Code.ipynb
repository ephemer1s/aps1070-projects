```
#2 Generate Errors
import math

x = [[1, 1], [1, 2], [1, 3], [1, 4], [1, 5]] #data
t = [1.5, 2.5, 3.5, 4.5, 5.5] #targets
theta = [0.1, 0.7] #initial weights

def simple_h(x, theta, t):
    J, L, y = [], [], []
    for n in range(len(x)):
        v = 0
        for p in range(len(x[0])):
            v = v + x[n][p]*theta[p]
        y.append(v)
        L.append(0.5*(y[n]-t[n])**2) # MSE error
    J.append(sum(L)/len(x))
    return (y, theta, J)

# train hypothesis
y, theta, J = simple_h(x, theta, t)
```

```
#3 Gradient Descent
import math

x = [[1, 1], [1, 2], [1, 3], [1, 4], [1, 5]] #data
t = [1.5, 2.5, 3.5, 4.5, 5.5] #targets
theta = [0.1, 0.1] #initial weights

def simple_h(x, theta, t, epochs = 100, learning = 0.015):
    J = [] #error
    for ii in range(epochs):
        L, y = [], []
        for n in range(len(x)):
            v = 0
            for p in range(len(x[0])):
                v = v + x[n][p]*theta[p]
            y.append(v)
            L.append(0.5*(y[n]-t[n])**2) # MSE error

            #gradient descent to compute new weights
            for p in range(len(theta)):
                d = x[n][p]*(y[n]-t[n])
                theta[p] = theta[p] - learning*d
        J.append(sum(L)/len(x))
    return (y, theta, J)

# train hypothesis
y, theta, J = simple_h(x, theta, t)
```
